```python
import numpy
import pandas
```

```python
# Series对象
s=pandas.Series([1,2,5,numpy.nan,6,8])
s
```

```python
# DataFrame对象
dates=pandas.date_range("20130101",periods=6) # 生成时间戳索引对象
# 以时间戳为指定的索引项，构造DataFrame，列名分别为ABCD，数据由numpy随机生成
df=pandas.DataFrame(numpy.random.randn(6,4),index=dates,columns=['A','B','C','D'])
df
```

```python
# 另一种构造DataFrame的方式
df2 = pandas.DataFrame({
    'A':1.0,
    'B':pandas.Timestamp("20130102"),
    'C':pandas.Series(1,index=list(range(4)),dtype="float32"),
    'D':numpy.array([3,3,3,3],dtype="int32"),
    'E':pandas.Categorical(["test","train","test","train"]),
    'F':"FOO"
})
df2
```

## Selection

```python
# 按字段选取
df["A"]
```

```python
# 按索引号选取
df[0:3]
```

```python
# 按索引值选取
df["20130102":"20130104"]
```

```python
# 使用loc方法选取指定索引值对应的一行
df.loc["20130101"]
```

```python
# 获取指定行列对应的区域，先行后列
df.loc[:,["A","B"]]
```

```python
df.loc["20130101","A"]
```

```python
# 等价于
df.at["20130101","A"]
```

```python
# 使用iloc方法，通过行列索引值，也是先行后列
df.iloc[3]
```

```python
df.iloc[1:3,2:4]
```

```python
df.iloc[[1,2,4],[0,2]]
```

```python
df.iloc[2,3]
```

```python
# 等价于
df.iat[2,3]
```

```python
# 按条件选择
df[df['A']>0]
```

```python
df2=df.copy()
df2["E"]=[2,3,4,3,2,5]
df2[df2["E"].isin([2,3])]
```

```python
# 赋值
df.at[dates[0],'A']=0
df.iat[0,1]=0.2
df.loc[:,'D']=[5]*len(df)
df
```

```python
df2=df.copy()
# 对大于0的单元格取负值
df2[df2>0]=-df2
df2
```
